```python
def SUM(A,n):
    if n==1:
        return A.sum(axis=0)
    else:
        return SUM(A.sum(axis=0),(n-1))

def ratioSUM(A,B,n):
    return SUM(A,n)/SUM(B,n)

def ratioS(A,B,string,n):
    sum_A = eval("SUM(A%s,n)" %string)
    sum_B = eval("SUM(B%s,n)" %string)
    return sum_A/sum_B
```
