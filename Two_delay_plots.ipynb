```python
from psana import *
import numpy as np
import os
import matplotlib.pyplot as plt
from skbeam.core.accumulators.histogram import Histogram
%matplotlib nbagg

runs = [75,77]
labels = ['probe_lead','overlap']
```

```python
folder_lead = "/reg/d/psdm/AMO/amolr2516/results/npzfiles/run00%d" %runs[0]
files_lead = os.listdir(folder_lead)

folder_overlap = "/reg/d/psdm/AMO/amolr2516/results/npzfiles/run00%d" %runs[1]
files_overlap = os.listdir(folder_overlap)
```

```python
lead_line = np.zeros(714)
overlap_line = np.zeros(714)
lead_var = np.zeros(714)
overlap_var = np.zeros(714)
normalisation = True

for fi in files_lead:
    data = np.load(os.path.join(folder_lead,fi))
    gas_mask = (np.nan_to_num(data['GasDetector'][:,:2]).mean(axis=1)>0.1)
    if normalisation:
        lead_line += (data['SHESwf'][gas_mask,:]/(data['GasDetector'][gas_mask,:2].mean(axis = 1))).mean(axis = 0)
        lead_var += (data['SHESwf'][gas_mask,:]/(data['GasDetector'][gas_mask,:2].mean(axis = 1))).var(axis = 0)
    else:
        lead_line += data['SHESwf'][gas_mask].mean(axis = 0)
        lead_var += data['SHESwf'][gas_mask].var(axis = 0)
        
for fi in files_overlap:
    data = np.load(os.path.join(folder_overlap,fi))
    gas_mask = (np.nan_to_num(data['GasDetector'][:,:2]).mean(axis=1)>0.1)
    if normalisation:
        overlap_line += (data['SHESwf'][gas_mask,:]/(data['GasDetector'][gas_mask,:2]).mean(axis = 1)).mean(axis = 0)
        overlap_var += (data['SHESwf'][gas_mask,:]/(data['GasDetector'][gas_mask,:2]).mean(axis = 1)).var(axis = 0)
    else:
        overlap_line += data['SHESwf'][gas_mask].mean(axis = 0)
        overlap_var += data['SHESwf'][gas_mask].var(axis = 0)
```

```python
colors = {'lead':'b', 'overlap':'r'}
plt.figure(figsize=(8,6))
plt.plot(range(714),lead_line,label = 'lead',color = colors['lead'])
plt.plot(range(714),overlap_line,label = 'overlap',color = colors['overlap'])
plt.fill_between(range(714), lead_line + np.sqrt(lead_var), lead_line - np.sqrt(lead_var),alpha = 0.3,color = colors['lead'])
plt.fill_between(range(714), overlap_line + np.sqrt(overlap_var), overlap_line - np.sqrt(overlap_var),alpha = 0.3,color = colors['overlap'])
plt.legend()
plt.title('Unnormalised')
```
