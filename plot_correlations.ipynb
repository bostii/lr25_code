```python
from psana import *
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
from skbeam.core.accumulators.histogram import Histogram
%matplotlib nbagg

run = 178
folder = "/reg/d/psdm/AMO/amolr2516/results/npzfiles/run%04d" %run
files = os.listdir(folder)
```

```python
gas = np.empty(0)
ph1 = np.empty(0)
ph2 = np.empty(0)
int1 = np.empty(0)
int2 = np.empty(0)
L3 = np.empty(0)
int_ges = np.empty(0)
itof_count = np.empty(0)
for file in files:
    data = np.load(os.path.join(folder,file))
    gas = np.hstack((gas,data['GasDetector'][:,:2].mean(axis = 1)))
    ph1 = np.hstack((ph1,data['UXSpc'][:,0]))
    ph2 = np.hstack((ph2,data['UXSpc'][:,3]))
    int1 = np.hstack((int1,data['UXSpc'][:,2]))
    int2 = np.hstack((int2,data['UXSpc'][:,5]))
    int_ges = np.hstack((int_ges,data['UXSpc'][:,2]+data['UXSpc'][:,5]))
    itof_count = np.hstack((itof_count,np.sum(data['ITOF'][:,:],axis=1)))
    L3 = np.hstack((L3,data['EBeamParameters'][:,5]))
    break #only one file
ph_en = np.vstack((ph1,int1,ph2,int2)).T
for ii in range(ph_en.shape[0]):
    if ph_en[ii,2]<ph_en[ii,0]:
        #ph_en[ii,:2] = ph_en[ii,:2:-1]
        #ph_en[ii,2:] = ph_en[ii,[3,2]]
        ph_en[ii,:] = np.roll(ph_en[ii,:],2)
others = np.vstack((gas,itof_count,L3,int_ges)).T 
DATA = np.hstack((ph_en,others))
mask = np.logical_and(gas>0.1,ph_en[:,0]>450)
mask = np.logical_and(mask,ph_en[:,0]<650)
#mask = np.logical_and(mask,int1<50000)
mask = np.logical_and(mask,ph_en[:,2]>650)
mask = np.logical_and(mask,ph_en[:,2]<850)
#mask = np.logical_and(mask,int1>5000)
#mask = np.logical_and(mask,int2>5000)
DATA = DATA[mask,:]
```

```python
df = pd.DataFrame(DATA,columns = ['ph1','int1','ph2','int2','gas','itof_count','L3','int_ges'],index=range(DATA.shape[0]))
```
