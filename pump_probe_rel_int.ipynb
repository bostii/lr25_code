```python
from psana import *
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.pylab as pl
from skbeam.core.accumulators.histogram import Histogram
%matplotlib nbagg

runs = [140,142,144,146,148,151,155]
```

```python
ratios = []
dists = []
for i in range(len(runs)):
    folder = "/reg/d/psdm/AMO/amolr2516/results/npzfiles/run%04d" %runs[i]
    files = os.listdir(folder)
    counter = 0
    run_ints = 0
    run_dist = np.empty(0)
    for fi in files:
        data = np.load(os.path.join(folder,fi))
        intensities = np.nan_to_num(np.vstack((data['UXSpc'][:,2],data['UXSpc'][:,5])).T)
        mask = (intensities>5000).all(axis = 1)  
        run_ints += (intensities[mask,0]/intensities[mask,1]).mean()
        run_dist = np.hstack((run_dist,intensities[mask,0]/intensities[mask,1]))
        counter +=1
    ratios.append(run_ints/counter)
    dists.append(run_dist)
plt.figure()
plt.plot(ratios)

```
