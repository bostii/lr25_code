```python
from psana import *
import numpy as np
import os
import matplotlib.pyplot as plt
from skbeam.core.accumulators.histogram import Histogram
%matplotlib nbagg

run = 75
folder = "/reg/d/psdm/AMO/amolr2516/results/npzfiles/run00%d" %run
files = os.listdir(folder)
```

```python
for fi in files:
    data = np.load(os.path.join(folder,fi))
    break

    data.keys()
```

```python
gas = []
pulse_COM = []
nan_count = 0
for i in range(data['GasDetector'].shape[0]):
    gas.append(data['GasDetector'][i,:2].mean())
    COM = data['UXSpc'][i,0]*data['UXSpc'][i,2] + data['UXSpc'][i,3]*data['UXSpc'][i,5]
    if np.isnan(COM):
        continue
    pulse_COM.append(0.5*COM)

gas = np.asarray(gas) 
pulse_COM = np.asarray(pulse_COM)
print gas.mean(), gas.std(), pulse_COM.mean(), pulse_COM.std()
```
