```python
from psana import *
import numpy as np
import os
import matplotlib.pyplot as plt
import math
from skbeam.core.accumulators.histogram import Histogram
%matplotlib nbagg
runs = [104,106,107,110,113,115]
```

```python
rejection = []
for i in range(len(runs)):
    folder = "/reg/d/psdm/AMO/amolr2516/results/npzfiles/run%04d" %runs[i]
    files = os.listdir(folder)
    counter = 0.0
    tot = 0.0
    for fi in files:
        data = np.load(os.path.join(folder,fi))
        energies = np.sort(np.vstack((data['UXSpc'][:,0],data['UXSpc'][:,3])).T)
        intensities = np.nan_to_num(np.vstack((data['UXSpc'][:,2],data['UXSpc'][:,5])).T)
        tot += intensities.shape[0]
        counter += (intensities[:,1]==0).sum()
        
    print 'Rejecting ' + str(100*round(counter/tot,2)) + ' percent of pump pulses in run %d' % runs[i]
```

```python
sum_int = []
gas = []
for i in range(2):
    folder = "/reg/d/psdm/AMO/amolr2516/results/npzfiles/run%04d" %runs[i]
    files = os.listdir(folder)
    for fi in files:
        data = np.load(os.path.join(folder,fi))
        gas_mask = data['GasDetector'][:,:2].mean(axis = 1) >0.3
        gas = np.hstack((gas,data['GasDetector'][gas_mask,:2].mean(axis=1)))
        sum_int = np.hstack((sum_int,data['UXSpc'][gas_mask,2]*data['UXSpc'][gas_mask,1] + data['UXSpc'][gas_mask,5]*data['UXSpc'][gas_mask,4]))
```

```python
plt.figure()
plt.scatter(gas[~np.isnan(sum_int)],sum_int[~np.isnan(sum_int)],alpha = 0.05)
plt.xlim(0.3,1.2)
plt.ylim(0,1e7)
np.corrcoef(gas[~np.isnan(sum_int)],sum_int[~np.isnan(sum_int)])
```
