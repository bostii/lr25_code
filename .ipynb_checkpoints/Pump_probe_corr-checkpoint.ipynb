```python
from psana import *
import numpy as np
import os
import matplotlib.pyplot as plt
from skbeam.core.accumulators.histogram import Histogram
%matplotlib nbagg

runs = [151]
labels = ['probe_lead','overlap']
```

```python
folder_lead = "/reg/d/psdm/AMO/amolr2516/results/npzfiles/run%04d" %runs[0]
files_lead = os.listdir(folder_lead)
```

```python
counter = 1
pump = []
probe = []
for fi in files_lead:
    if counter == 20:
        break
    data = np.load(os.path.join(folder_lead,fi))
    energies = np.sort(np.vstack((data['UXSpc'][:,0],data['UXSpc'][:,3])).T)
    gas = data['GasDetector'][:,:2].mean(axis = 1)
    intensities = np.nan_to_num(np.vstack((data['UXSpc'][:,2],data['UXSpc'][:,5])).T)
    mask = (intensities>500).all(axis = 1)  
    counter +=1
```

```python
plt.figure()
plt.hist(intensities[:,0],bins = np.arange(1,200000,500))
plt.xlim(0,200000)
plt.title('Probe intensity distribution')
```

```python
plt.figure()
plt.hist(intensities[:,1],bins = np.arange(1,200000,500))
plt.title('Pump intensity distribution')
plt.xlim(0,200000)
```

```python
plt.figure()
plt.scatter(energies[mask,0],energies[mask,1],alpha = 0.5)
plt.xlim(450,650)
plt.ylim(650,850)
```
